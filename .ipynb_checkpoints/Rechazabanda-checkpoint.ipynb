# Diseño de Filtro FIR Rechaza Banda en Python

## Introducción

Los filtros FIR (Finite Impulse Response) son filtros digitales que tienen una respuesta de impulso de duración finita. Estos filtros se caracterizan por su estabilidad y por tener una respuesta lineal de fase, lo cual los hace ideales para muchas aplicaciones en procesamiento de señales.

En este notebook, diseñamos un filtro FIR rechaza banda utilizando Python. Un filtro rechaza banda es un filtro que elimina o atenúa las frecuencias dentro de un rango específico, permitiendo que las demás frecuencias pasen sin cambios significativos.

## Teoría del Filtro FIR Rechaza Banda

### Ecuaciones Fundamentales

La ecuación de un filtro FIR de orden \( M-1 \) se puede expresar como:

$$
y[n] = \sum_{k=0}^{M-1} h[k] x[n-k]
$$

donde:
- \( y[n] \) es la señal de salida.
- \( x[n] \) es la señal de entrada.
- \( h[k] \) son los coeficientes del filtro.
- \( M \) es el número de coeficientes del filtro (también conocido como la longitud del filtro).

### Diseño del Filtro

Para diseñar un filtro FIR rechaza banda, utilizamos la técnica de ventanas, donde aplicamos una ventana a la respuesta de impulso ideal del filtro. La respuesta de impulso ideal para un filtro rechaza banda se define como:

$$
H_d(e^{j\omega}) = \begin{cases} 
1 & \text{si } \omega \leq \omega_{c1} \text{ o } \omega \geq \omega_{c2} \\
0 & \text{si } \omega_{c1} < \omega < \omega_{c2}
\end{cases}
$$

donde \( \omega_{c1} \) y \( \omega_{c2} \) son las frecuencias de corte normalizadas.

La implementación práctica de esta respuesta ideal se logra mediante la convolución con una ventana de Hamming:

$$
w[n] = 0.54 - 0.46 \cos \left( \frac{2\pi n}{M-1} \right)
$$

Finalmente, los coeficientes del filtro FIR se calculan como:

$$
h[n] = h_d[n] \cdot w[n]
$$

## Parámetros del Filtro

Definimos los parámetros básicos para nuestro filtro FIR rechaza banda:

- `lowcut`: Frecuencia de corte inferior (en Hz).
- `highcut`: Frecuencia de corte superior (en Hz).
- `nyq_rate`: Frecuencia de Nyquist (la mitad de la frecuencia de muestreo).
- `M`: Número de coeficientes del filtro.

## Frecuencia de Muestreo

La frecuencia de muestreo \( f_s \) es la frecuencia a la cual se toman las muestras de una señal continua para convertirla en una señal digital. Según el teorema de Nyquist, la frecuencia de muestreo debe ser al menos el doble de la frecuencia más alta presente en la señal que se desea muestrear para evitar el aliasing.

$$
f_s \geq 2 \cdot f_{\text{max}}
$$

```python
lowcut = 500.0  # Frecuencia de corte inferior (Hz)
highcut = 1500.0  # Frecuencia de corte superior (Hz)
nyq_rate = fs / 2.0
M = 101  # Número de coeficientes del filtro
```

```python
# Leer el archivo de audio
rate, audio = wavfile.read('audio.wav')

# Mostrar la frecuencia de muestreo
print(f'Frecuencia de muestreo: {rate} Hz')

# Mostrar la duración del audio
duration = len(audio) / rate
print(f'Duración del audio: {duration} segundos')
```

```python
fs = 44100  # Frecuencia de muestreo (Hz)
```

## Diseño del Filtro

Usamos la función `firwin` de `scipy.signal` para diseñar el filtro FIR rechaza banda. Esta función genera los coeficientes del filtro basado en los parámetros definidos.

### Descripción del Código

- **firwin**: Es una función de `scipy.signal` que diseña filtros FIR utilizando la técnica de ventanas.
- **M**: Es el número de coeficientes del filtro. A mayor número de coeficientes, mejor será la precisión del filtro, pero a costa de un mayor costo computacional.
- **[lowcut / nyq_rate, highcut / nyq_rate]**: Especifica las frecuencias de corte del filtro en términos de la frecuencia de Nyquist. La frecuencia de Nyquist es la mitad de la frecuencia de muestreo.
  - **lowcut**: Frecuencia de corte inferior en Hz.
  - **highcut**: Frecuencia de corte superior en Hz.
  - **nyq_rate**: Frecuencia de Nyquist (la mitad de la frecuencia de muestreo).
- **pass_zero=True**: Especifica que el filtro es un filtro rechaza banda. Al usar `True`, estamos diciendo que el filtro debe tener una ganancia de cero en las bandas de paso.
- **window='hamming'**: Especifica que se utilizará una ventana de Hamming para el diseño del filtro. Las ventanas se usan para controlar el ripple en la respuesta en frecuencia del filtro.




```python
taps_bp = firwin(M, [lowcut / nyq_rate, highcut / nyq_rate], pass_zero=True, window='hamming')
print("Coeficientes del filtro FIR rechaza banda:")
print(taps_bp)
```

```python
# Graficar los coeficientes del filtro
plt.figure(figsize=(15, 8))
plt.stem(taps_bp)
plt.title("Coeficientes del Filtro FIR Rechaza Banda")
plt.xlabel("Índice del Coeficiente")
plt.ylabel("Valor del Coeficiente")
plt.grid()
plt.show()
```

### Observación de la Gráfica de Coeficientes del Filtro FIR

#### Simetría

Los coeficientes son  simétricos alrededor del punto central. Esto es una confirmación visual de que el filtro FIR tiene una respuesta en fase lineal. Los valores de los coeficientes en el lado izquierdo de la gráfica son iguales a los valores de los coeficientes en el lado derecho, respecto al coeficiente central.

#### Coeficiente Central

El coeficiente central tiene un valor significativamente mayor que los demás coeficientes. Este valor representa la contribución principal del filtro en el centro de la banda rechazada.

#### Amplitud de los Coeficientes

Los coeficientes en los extremos de la gráfica tienen valores cercanos a cero, lo cual es típico en el diseño de filtros FIR con ventanas. Esto minimiza las discontinuidades en los bordes del filtro, ayudando a reducir el ripple en la respuesta en frecuencia.


Respuesta en Frecuencia del Filtro
Utilizamos la función freqz para calcular la respuesta en frecuencia del filtro FIR diseñado.

```python
w, h = freqz(taps_bp, worN=8000)
plt.figure(figsize=(15, 8))
plt.plot(0.5 * fs * w / np.pi, np.abs(h))
plt.title("Respuesta en Frecuencia del Filtro Rechaza Banda")
plt.xlabel("Frecuencia (Hz)")
plt.ylabel("Ganancia")
plt.grid()
plt.show()
```

### Observaciones de la Gráfica

- La gráfica muestra una caída pronunciada en la ganancia cerca de la frecuencia de corte inferior y una ganancia muy baja en un rango de frecuencias. Esto indica que el filtro atenúa efectivamente las frecuencias dentro de la banda rechazada.

- Fuera de la banda rechazada, la ganancia es cercana a 1, indicando que el filtro permite el paso de las frecuencias fuera de la banda rechazada sin mucha atenuación.


### Aplicación del Filtro a la señal de audio

### Argumentos de `lfilter`

- **taps_bp**: Este es el primer argumento que representa los coeficientes del numerador del filtro FIR. Estos coeficientes se calculan previamente utilizando la función `firwin` y definen las características del filtro, como las frecuencias de corte y el tipo de ventana utilizada.
- **1.0**: Este es el segundo argumento que representa los coeficientes del denominador del filtro. Para un filtro FIR, el denominador es simplemente `1.0` porque el filtro FIR no tiene una respuesta dependiente de su propia salida (a diferencia de un filtro IIR). Esto significa que el filtro es una respuesta de impulso finita.
- **audio**: Este es el tercer argumento que representa la señal de entrada que queremos filtrar. En este caso, es la señal de audio que se ha leído del archivo `audio.wav`.

### Proceso de Filtrado

La función `lfilter` toma los coeficientes del filtro y la señal de entrada, y aplica la operación de convolución para obtener la señal filtrada. La convolución se realiza entre los coeficientes del filtro y la señal de entrada, lo que resulta en una nueva señal (`filtered_x_bp`) que tiene las frecuencias especificadas atenuadas o eliminadas.


```python
# Leer la señal de audio 
rate, audio = wavfile.read('audio.wav')
audio = audio.astype(np.float32)

# Aplicar el filtro a la señal
filtered_x_bp = lfilter(taps_bp, 1.0, audio)

# Visualizar las señales original y filtrada
inicio = 0  # Índice de inicio para la visualización
fin = len(audio)  # Índice de fin para la visualización
t = np.arange(inicio, fin) / fs  # Vector de tiempo


# Calcular el retardo del filtro
# El retardo del filtro (delay_bp) se calcula como la mitad del número de coeficientes (M) menos uno, dividido por la frecuencia de muestreo (fs).
# Esto se hace porque un filtro FIR con coeficientes simétricos introduce un retardo de grupo constante, que es la mitad del orden del filtro.
delay_bp = 0.5 * (M - 1) / fs

# Crear una nueva figura para la gráfica
plt.figure(figsize=(15, 8))

# Graficar la señal original
# Se grafica la señal original en azul para que sirva como referencia.
plt.plot(t[inicio:fin], audio[inicio:fin], label='Original')

# Graficar la señal filtrada
# Se grafica la señal filtrada en rojo. La señal filtrada se desplaza hacia la izquierda por el valor del retardo calculado (delay_bp)
# para alinearla temporalmente con la señal original.
plt.plot(t[inicio:fin] - delay_bp, filtered_x_bp[inicio:fin], 'r-', label='Filtrado')

# Añadir etiquetas a los ejes X e Y
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')

# Añadir un título a la gráfica
plt.title('Señal Original y Filtrada (Rechaza Banda)')

# Añadir una cuadrícula a la gráfica para mejorar la legibilidad
plt.grid()

# Añadir una leyenda para identificar las señales original y filtrada
plt.legend()

# Mostrar la gráfica
plt.show()
```

### Comparación en el Dominio de la Frecuencia

Para evaluar cómo el filtro FIR rechaza banda afecta a la señal de audio en términos de su contenido frecuencial, calculamos y graficamos la Transformada de Fourier (FFT) de ambas señales: la señal original y la señal filtrada. Esta comparación nos permite observar cómo se modifican las componentes de frecuencia debido a la aplicación del filtro.



```python
# Calculamos la Transformada de Fourier de ambas señales
audio_fft = np.fft.fft(audio)
filtered_x_bp_fft = np.fft.fft(filtered_x_bp)

# Frecuencias correspondientes
frequencies = np.fft.fftfreq(len(audio), 1/fs)

# Graficamos las transformadas
plt.figure(figsize=(15, 8))
plt.plot(frequencies[:len(frequencies)//2], np.abs(audio_fft)[:len(frequencies)//2], label='Original')
plt.plot(frequencies[:len(frequencies)//2], np.abs(filtered_x_bp_fft)[:len(frequencies)//2], label='Filtrada')
plt.xlabel('Frecuencia (Hz)')
plt.ylabel('Magnitud')
plt.title('Transformada de Fourier de la Señal Original y Filtrada')
plt.grid()
plt.legend()
plt.show()
```


### Efecto General del Filtro

En la parte baja del espectro de frecuencias, se observa una disminución drástica de la magnitud en la señal filtrada. Esto confirma que el filtro está eliminando componentes no deseadas en la banda específica mientras preserva las demás.

### Frecuencias Altas

En las frecuencias más altas (por encima de 1500 Hz), la gráfica muestra que las magnitudes de las componentes frecuenciales permanecen casi iguales entre la señal original y la filtrada, lo que confirma que estas frecuencias no han sido afectadas por el filtro.

### Evaluación del Rendimiento del Filtro

Calculamos el coeficiente de correlación entre la señal original y la señal filtrada para evaluar el rendimiento del filtr.4f}')


```python
correlation_bp = np.corrcoef(audio, filtered_x_bp)[0, 1]
print(f'Coeficiente de correlación (Rechaza Banda): {correlation_bp:.4f}')
```

### Interpretación del Coeficiente de Correlación

El coeficiente de correlación es una medida que indica el grado de similitud entre la señal original y la señal filtrada. Su valor oscila entre -1 y 1:

- Un coeficiente de correlación de 1 indica una correlación perfecta, lo que significa que las dos señales son idénticas.
- Un coeficiente de correlación de 0 indica que no hay correlación, es decir, las dos señales no tienen ninguna relación lineal.
- Un coeficiente de correlación de -1 indica una correlación perfecta negativa, lo que significa que las dos señales son opuestas.

En este caso, un coeficiente de correlación de 0.5238 indica una correlación moderada positiva entre la señal original y la señal filtrada. Esto sugiere que el filtro ha modificado la señal original de manera significativa, lo cual es esperado ya que el filtro rechaza banda está diseñado para eliminar componentes de frecuencia específicas (500 Hz a 1500 Hz) mientras preserva las demás.
