Proyecto de Análisis y Filtrado de Señales de Audio de Instrumentos Musicales

Importación de Librerías y Configuración Inicial

```python
from matplotlib.pylab import *
import scipy.signal as signal
get_ipython().run_line_magic('matplotlib', 'inline')
import math
import numpy as np
import scipy.signal as sp
import matplotlib.pyplot as plt
import pandas as pd
import scipy.io.wavfile as waves
from IPython.display import Audio
from scipy.io.wavfile import read, write
```

Paso 1: Cargar y Preprocesar el Archivo de Audio

```python
# Cargar el archivo de audio
fs, x = read('audio.wav')  # Sustituir 'audio.wav' con la ruta de tu archivo de audio

# Verificar y manejar si los datos son estéreo (2 canales)
if len(x.shape) > 1:
    x = x[:, 0]  # Usar solo el primer canal si el audio es estéreo

# Normalizar la señal
x = x / float(np.max(np.abs(x)))

# Eliminar nivel DC
nivel_DC = np.mean(x)
x -= nivel_DC

# Crear el vector de tiempo
t = np.arange(0, float(len(x)) / fs, 1.0 / fs)

# Gráfica de la señal
plt.figure(figsize=(15, 8))
plt.plot(t, x)
plt.grid()
plt.title('Señal de Audio Original')
plt.ylabel('Amplitud', fontsize=18)
plt.xlabel('Tiempo', fontsize=18)
plt.show()

# Reproducir el audio
display(Audio(x, rate=fs))
```

Paso 2: Seleccionar un Segmento de 15 Segundos

```python
# Duración total del audio en segundos
duracion_total = len(x) / fs
# Duración del segmento a recortar (15 segundos)
duracion_segmento = 15
# Número de muestras para el segmento de 15 segundos
muestras_segmento = int(duracion_segmento * fs)
# Inicio del segmento (en muestras)
inicio = int((duracion_total - duracion_segmento) / 2 * fs)
# Final del segmento (en muestras)
fin = inicio + muestras_segmento
# Segmento de audio recortado
audio_recortado = x[inicio:fin]

# Crear el vector de tiempo para el segmento recortado
t_recortado = t[inicio:fin]

# Gráfica de la señal recortada
plt.figure(figsize=(15, 8))
plt.plot(t_recortado, audio_recortado)
plt.grid()
plt.title('Señal de Audio Recortada (15 segundos)')
plt.ylabel('Amplitud', fontsize=18)
plt.xlabel('Tiempo', fontsize=18)
plt.show()

# Reproducir el audio recortado
display(Audio(audio_recortado, rate=fs))
```

Paso 3: Análisis en Frecuencia

```python
# FFT del segmento recortado
N = len(audio_recortado)
T = 1.0 / fs
yf = np.fft.fft(audio_recortado)
xf = np.linspace(0.0, 1.0 / (2.0 * T), N // 2)

# Plot de la FFT
plt.figure(figsize=(12, 6))
plt.plot(xf, 2.0 / N * np.abs(yf[:N // 2]))
plt.grid()
plt.title('FFT de la Señal de Audio Recortada')
plt.xlabel('Frecuencia (Hz)')
plt.ylabel('Amplitud')
plt.show()

# Espectrograma del segmento recortado
plt.figure(figsize=(12, 6))
plt.specgram(audio_recortado, Fs=fs)
plt.title('Espectrograma de la Señal de Audio Recortada')
plt.xlabel('Tiempo (s)')
plt.ylabel('Frecuencia (Hz)')
plt.colorbar(label='Intensidad (dB)')
plt.show()
```

Paso 4: Diseño y Aplicación de Filtros

Teoría y Cálculo de Coeficientes de Filtros

Filtro Pasa Bajo Butterworth

```python
import numpy as np
from scipy.signal import butter

# Frecuencia de muestreo
fs = 44100.0  # Ejemplo: 44.1 kHz
# Frecuencia de corte para el filtro pasa bajo
fc_low = 1000.0  # 1 kHz
# Orden del filtro
order = 5

# Frecuencia de Nyquist
nyquist = 0.5 * fs
# Frecuencia de corte normalizada
normal_cutoff = fc_low / nyquist

# Cálculo de los coeficientes del filtro pasa bajo
b_low, a_low = butter(order, normal_cutoff, btype='low', analog=False)

print("Coeficientes b (numerador):", b_low)
print("Coeficientes a (denominador):", a_low)
```

Filtro Pasa Alto Butterworth

```python
# Frecuencia de corte para el filtro pasa alto
fc_high = 5000.0  # 5 kHz

# Frecuencia de corte normalizada
normal_cutoff = fc_high / nyquist

# Cálculo de los coeficientes del filtro pasa alto
b_high, a_high = butter(order, normal_cutoff, btype='high', analog=False)

print("Coeficientes b (numerador):", b_high)
print("Coeficientes a (denominador):", a_high)
```

Filtro Pasa Banda Butterworth

```python
# Frecuencias de corte para el filtro pasa banda
fc_low_band = 1000.0  # 1 kHz
fc_high_band = 5000.0  # 5 kHz

# Frecuencias de corte normalizadas
normal_cutoff_band = [fc_low_band / nyquist, fc_high_band / nyquist]

# Cálculo de los coeficientes del filtro pasa banda
b_band, a_band = butter(order, normal_cutoff_band, btype='band', analog=False)

print("Coeficientes b (numerador):", b_band)
print("Coeficientes a (denominador):", a_band)
```

```python
from scipy.signal import lfilter
from numpy import corrcoef

# Aplicar los filtros a la señal recortada
low_passed = lfilter(b_low, a_low, audio_recortado)
high_passed = lfilter(b_high, a_high, audio_recortado)
band_passed = lfilter(b_band, a_band, audio_recortado)

# Visualización de las señales filtradas
plt.figure(figsize=(15, 12))
plt.subplot(4, 1, 1)
plt.plot(t_recortado, audio_recortado)
plt.title('Señal Original')
plt.subplot(4, 1, 2)
plt.plot(t_recortado, low_passed)
plt.title('Señal Filtrada - Pasa Bajo')
plt.subplot(4, 1, 3)
plt.plot(t_recortado, high_passed)
plt.title('Señal Filtrada - Pasa Alto')
plt.subplot(4, 1, 4)
plt.plot(t_recortado, band_passed)
plt.title('Señal Filtrada - Pasa Banda')
plt.tight_layout()
plt.show()

# Reproducir las señales filtradas
display(Audio(low_passed, rate=fs))
display(Audio(high_passed, rate=fs))
display(Audio(band_passed, rate=fs))

# Calcular el coeficiente de correlación
corr_low = corrcoef(audio_recortado, low_passed)[0, 1]
corr_high = corrcoef(audio_recortado, high_passed)[0, 1]
corr_band = corrcoef(audio_recortado, band_passed)[0, 1]

print("Coeficiente de correlación (Pasa Bajo):", corr_low)
print("Coeficiente de correlación (Pasa Alto):", corr_high)
print("Coeficiente de correlación (Pasa Banda):", corr_band)
```

```python
from numpy import corrcoef

# Calcular el coeficiente de correlación
corr_low = corrcoef(audio_recortado, low_passed)[0, 1]
corr_high = corrcoef(audio_recortado, high_passed)[0, 1]
corr_band = corrcoef(audio_recortado, band_passed)[0, 1]

# Mostrar coeficientes de correlación
print(f"Coeficiente de correlación - Pasa Bajo: {corr_low}")
print(f"Coeficiente de correlación - Pasa Alto: {corr_high}")
print(f"Coeficiente de correlación - Pasa Banda: {corr_band}")

# Análisis en frecuencia para señales filtradas
yf_low = np.fft.fft(low_passed)
yf_high = np.fft.fft(high_passed)
yf_band = np.fft.fft(band_passed)

plt.figure()
plt.subplot(3, 1, 1)
plt.plot(xf, 2.0 / N * np.abs(yf_low[:N // 2]))
plt.title('FFT - Señal Pasa Bajo')
plt.grid()
plt.subplot(3, 1, 2)
plt.plot(xf, 2.0 / N * np.abs(yf_high[:N // 2]))
plt.title('FFT - Señal Pasa Alto')
plt.grid()
plt.subplot(3, 1, 3)
plt.plot(xf, 2.0 / N * np.abs(yf_band[:N // 2]))
plt.title('FFT - Señal Pasa Banda')
plt.grid()
plt.tight_layout()
plt.show()
```

Las señales filtradas se comparan con la señal original utilizando el coeficiente de correlación, que mide la similitud entre las dos señales. Este valor varía entre 0 y 1, donde 1 indica una correlación perfecta. Además, se realiza un análisis en frecuencia (FFT) de las señales filtradas para visualizar cómo se han modificado las componentes frecuenciales.
